**Q1**

$x_1, ..., x_n \sim Pois(\lambda)$

$\hat \lambda = \frac{1}{n} \sum_{i=1}^n x_i$

$bias(\hat \lambda) = \mathbb{E}_{\lambda}(\hat \lambda) - \lambda = \frac{1}{n} \sum_{i=1}^n \mathbb{E}_{\lambda}(x_i) - \lambda = \frac{n}{n} \lambda - \lambda = 0$

$\mathbb{V}(\hat \lambda) = \mathbb{V}(\frac{1}{n} \sum_{i=1}^n x_i) = \frac{1}{n^2} \sum_{i=1}^n \mathbb{V}(x_i) = \frac{n \lambda }{n^2} = \frac{ \lambda }{n} $

$se(\hat \lambda) = \sqrt{\mathbb{V}(\hat \lambda)} = \sqrt{\frac{\lambda}{n}}$

$MSE = \mathbb{E}[(\lambda - \hat \lambda)^2] = bias^2(\hat \lambda) + \mathbb{V}(\hat \lambda) = 0 + \frac{\lambda}{n} = \frac{\lambda}{n}$

**Q2**

$\hat \theta = 2 \bar X_n$

$bias(\hat \theta) = \mathbb{E}_{\theta}(\hat \theta) - \theta = \mathbb{E}_{\theta}(2 \bar X_n) - \theta = \mathbb{E}_{\theta}(2 \frac{1}{n} \sum_{i=1}^n x_i) - \theta = \frac{2}{n} \sum_{i=1}^n \mathbb{E}_{\theta}(x_i) - \theta = \frac{2 \cdot n}{n} \frac{\theta}{2} - \theta = 0$


$\mathbb{V}(\hat \theta) = \mathbb{V}(\frac{2}{n} \sum_{i=1}^n x_i) = \frac{4}{n^2} \sum_{i=1}^n \mathbb{V}(x_i) = \frac{4n}{n^2} \frac{\theta^2}{12} = \frac{\theta^2}{3 n}$

$se(\hat \theta) = \sqrt{\mathbb{V}(\hat \theta)} = \sqrt{\frac{1}{3n}}\theta$


$MSE = \mathbb{E}[(\theta - \hat \theta)^2] = bias(\hat \theta)^2 + \mathbb{V}(\hat \theta) = 0 + \frac{\theta^2}{3n} = \frac{\theta^2}{3n}$


## P&P Ex 2

$\hat F_n(x) = \frac{\sum_{i=1}^n I(X_i \le x)}{n}$

, where $I(X_i \le x) = \left\{
\begin{array}{l}
1,\quad X_i \le x \\
0,\quad X_i > x
\end{array}
\right.$

**a)** $\mathbb{E} (\hat F_n(x)) = F(x) - ?$

$\mathbb{E} (\hat F_n(x)) = \mathbb{E} [\frac{\sum_{i=1}^n I(X_i  \le x)}{n}] =  \frac{\sum_{i=1}^n \mathbb{E} [I(X_i  \le x)]}{n} = \frac{\sum_{i=1}^n P(X_i \le x)}{n} = \frac{\sum_{i=1}^n F(x)}{n} =  \frac{n F(x)}{n} = F(x)$

Because:

As $\mathbb{E}(X) = \int x\, dF(x)$, and $I(X_i  \le x)=1$ when $X_i < x$:

+ $\mathbb{E}[I(X_i  \le x)] = 1 \cdot P(I(X_i  \le x)=1) + 0 \cdot P(I(X_i  \le x)=0) = P(I(X_i  \le x)=1) = P(X_i  \le x)$

**b)** $\mathbb{V} (\hat F_n(x)) = \frac{F(x)(1 - F(x))}{n} - ?$

$\mathbb{V} (\hat F_n(x)) = \mathbb{V}(\frac{\sum_{i=1}^n I(X_i \le x)}{n}) = \frac{1}{n^2} \sum_{i=1}^n \mathbb{V}(I(X_i \le x)) = \frac{1}{n^2} \sum_{i=1}^n [P(X_i \le x) - P^2(X_i \le x)] = \frac{1}{n^2} \sum_{i=1}^n F(x)(1 - F(x)) = \frac{1}{n} F(x)(1 - F(x))$

Because:

+ $\mathbb{V}[I(X_i  \le x)] = \mathbb{E}[I^2(X_i \le x)] - \mathbb{E}[I(X_i \le x)]^2 = \mathbb{E}[I^2(X_i \le x)] - P^2(X_i \le x)$
+ $\mathbb{E}[I^2(X_i \le x)] = 1 \cdot P(I^2(X_i \le x) = 1) + 0 \cdot P(I^2(X_i \le x) = 0) = P(X_i \le x)$

**c)** $MSE = \frac{1}{n} F(x)(1 - F(x))$

+ $MSE = bias^2(\hat F_n(x)) + \mathbb{V}(\hat F_n(x)) = 0 + \mathbb{V}(\hat F_n(x)) = \frac{1}{n} F(x)(1 - F(x))$
+ $bias(\hat F_n(x)) = \mathbb{E}(\hat F_n) - F(x) = 0$

## Programming Ex 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('future.no_silent_downcasting', True)
```

```python
rng = np.random.default_rng(seed=123)
```

```python
def compute_cdf(x):
    cnt = 0
    n = len(x)
    cdf = np.zeros(n)
    min_x = min(x)
    max_x = max(x)
    
    x = sorted(x)
    for i, x_i in enumerate(x):
        
        cnt += 1
        cdf[i] += cnt
    cdf /= n
    return cdf, x
```

```python
def bootstrapping(x, sizes):  
    n = len(x)
    Tboot = []
    for B_i in sizes:
        Tboot.append([])
        for i in range(B_i):
            Tboot[-1].append(np.median(rng.choice(x, size=n, replace=True)).item())
    return Tboot
```

```python
def get_conf(Tboot, B):
    conf_left = []
    conf_right = []
    alpha = 1 - 0.95
    for i in range(len(B)):
        conf_left.append(np.quantile(Tboot[i], alpha/2))
        conf_right.append(np.quantile(Tboot[i], 1 - alpha/2))
    return conf_left, conf_right
```

```python
df = pd.read_excel('./DeathsByState_2019_2020.xlsx', )
```

```python
df.replace('X', 0, inplace=True)
```

```python
n_deaths_2019 = df[df['Jahr'] == 2019].drop(columns=['Jahr', 'Bundesland', '29.02.']).sum(0).values
n_deaths_2020 = df[df['Jahr'] == 2020].drop(columns=['Jahr', 'Bundesland']).sum(0).values

date = df.columns[2:]
```

```python
cdf_2019, x_2019 = compute_cdf(n_deaths_2019)
cdf_2020, x_2020 = compute_cdf(n_deaths_2020)
```

```python
plt.plot(x_2019, cdf_2019, label='2019', color='blue')
plt.vlines(x=x_2019, ymin=0, ymax=0.1, color='blue', linestyles='solid', lw=1, alpha=0.5)

plt.plot(x_2020, cdf_2020, label='2020', color='orange')
plt.vlines(x=x_2020, ymin=0, ymax=0.1, color='orange', linestyles='solid', lw=1, alpha=0.5)
plt.title('CDF of daily deaths in Germany')

plt.legend()
plt.plot()
```

```python
med_2019 = np.median(n_deaths_2019)
med_2020 = np.median(n_deaths_2020)
```

```python
B = [100, 500, 1000]
```

```python
Tboot2019 = bootstrapping(n_deaths_2019, B)
```

```python
Tboot2020 = bootstrapping(n_deaths_2020, B)
```

```python
conf2019_left, conf2019_right = get_conf(Tboot2019, B)
```

```python
conf2020_left, conf2020_right = get_conf(Tboot2020, B)
```

```python
with open('output.txt', 'w') as f:
    for i in range(len(B)):
        line = f'{B[i]}, {conf2019_left[i]}, {conf2019_right[i]}, {conf2020_left[i]}, {conf2020_right[i]}\n'
        f.write(line)
```
