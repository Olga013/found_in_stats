## Pen and paper ex.3

$= \int_0^y \int_0^1 y^2 \, dx dy + \int_0^x \int_0^1 x^2 \, dx dy - \frac{4}{9}= \int_0^1 y^3 dy + \int_0^1 x^3 dx - \frac{4}{9}= \frac{2}{4} - \frac{4}{9} = \frac{18 - 16}{36} = \frac{1}{18}$

## Pen and paper ex.5

$f_{X,Y}(x, y) = \left\{

\frac{1}{3} (x + y),\quad 0\le x \le 1, \quad 0\le y \le 2, \\

\end{array}
\right.$

$V(2X - 3Y + 8) = ?$

As $V(\sum_i a_i X_i) = \sum_i a^2_i V(X_i) + 2 \sum \sum_{i < j} a_i a_j Cov(X_i, X_j)$:

$V(2X - 3Y + 8) = 4 V(X) + 9 V(Y) + 0 - 2 \cdot 6 \cdot Cov(X, Y) = 4 V(X) + 9 V(Y) - 12 \cdot Cov(X, Y)$

Also, 

$Cov(X, Y) = \mathbb{E}(X \cdot Y) - \mathbb{E}(X)\cdot \mathbb{E}(Y)$

<br>

$\underline{\mathbb{E}(X \cdot Y)} = \int_0^1 \int_0^2 x \cdot y \frac{1}{3} (x + y)\, dx dy = \int_0^1 \int_0^2 \frac{1}{3} (x^2 y + x y^2)\, dx dy = \frac{1}{3} \int_0^1 dx (\frac{x^2 y^2}{2} + {x y ^ 3}{3}) |^2_0 = \frac{1}{3} \int_0^1 (2 x^2 + \frac{8}{3} x) dx = \frac{1}{3}(\frac{2 x ^3}{3} + \frac{x^2 8}{3 \cdot 2})|_0^1 = \frac{1}{3}(\frac{2}{3} + \frac{8}{6}) =$ 
$= \frac{1}{3}(\frac{2}{3}+\frac{4}{3}) = \underline{\frac{2}{3}}$

<br>

$f_X(x) = \int_0^2 \frac{1}{3} (x + y)\, dy = \frac{1}{3} (x y + \frac{y^2}{2})|_0^2 = \frac{1}{3} (2 x + 2) = \frac{2}{3} (x + 1)$

$\underline{\mathbb{E}(X)} = \int x f_X(x)\, dx = \frac{2}{3} \int_0^1 x (x + 1)\, dx = \frac{2}{3} (\frac{x ^ 3}{3} + \frac{x^2}{2})|_0^1 = \frac{2}{3} (\frac{1}{3} + \frac{1}{2}) = \frac{2}{3} (\frac{2}{6} + \frac{3}{6}) = \frac{2}{3} \frac{5}{6} = \underline{\frac{5}{9}}$

<br>

$f_Y(y) = \int_0^1 \frac{1}{3} (x + y)\, dx = \frac{1}{3} (y + \frac{1}{2})$

$\underline{\mathbb{E}(Y)} = \int y f_Y(y)\, dy = \frac{1}{3} \int_0^2 (y^2 + \frac{y}{2})\, dy = \frac{1}{3} (\frac{y ^ 3}{3} + \frac{y^2}{4})|_0^2 = \frac{1}{3} (\frac{8}{3} + 1) = \underline{\frac{11}{9}}$

<br>

As $V(X) = \mathbb{E}(X^2) - \mathbb{E^2}(X)$, we need to find $\mathbb{E}(X^2)$:

$\underline{\mathbb{E}(X^2)} = \int x^2 \cdot f_X(x)\, dx = \frac{2}{3} \int_0^1 x^2 (x + 1) dx = \frac{2}{3} (\frac{x^4}{4} + \frac{x^3}{3}) |^1_0 = \frac{2}{3} (\frac{1}{4} + \frac{1}{3}) = \frac{2}{3} (\frac{3+4}{12}) = \underline{\frac{7}{18}}$

<br>

The same for $V(Y) = \mathbb{E}(Y^2) - \mathbb{E^2}(Y)$, we need to find $\mathbb{E}(Y^2)$:

$\underline{\mathbb{E}(Y^2)} = \int y^2 \cdot f_Y(y)\, dy = \frac{1}{3} \int_0^2 y^2 (y + \frac{1}{2}) dy = \frac{1}{3} (\frac{y^4}{4} + \frac{y^3}{3 \cdot 2}) |^2_0 = \frac{1}{3} (\frac{16}{4} + \frac{8}{6}) = \frac{1}{3} (\frac{48}{12} + \frac{16}{12}) = \frac{64}{12 \cdot 3} = \frac{32}{6 \cdot 3} = \underline{\frac{16}{9}}$

$V(2X - 3Y + 8) =  4 (\mathbb{E}(X^2) - \mathbb{E^2}(X)) + 9 (\mathbb{E}(Y^2) - \mathbb{E^2}(Y)) - 12 \cdot (\mathbb{E}(X \cdot Y) - \mathbb{E}(X)\cdot \mathbb{E}(Y)) = $

## Programming ex. 1

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#i.)
## IMPORT DATA FROM Input.txt
inputdata = np.loadtxt('./data/input/Input.txt')
#print(inputdata)
inputdata = [int(item) for item in inputdata]

## SET RANDOM SEED
np.random.seed(inputdata[0])
```

```python
N = np.array(inputdata[1:])
lam = 1
k = 10
```

```python
#calculate mean and var
mean = np.zeros((len(N), k))
var = np.zeros((len(N)))
for j in range(len(N)):
    for i in range(k):
        mean[j, i] = np.random.poisson(lam=1.0, size=N[j]).mean()
    #we consider sample variance here where the sample consists of a set of mean(X_n)
    var[j] = mean[j].var(ddof=1)
```

```python
#write into the file
with open('./data/output/MeanVar.txt', 'w') as f:
    for j in range(len(N)):
        line = ','.join(['%1.2f' % item for item in mean[j]]) + '\n'
        f.write(line)
    line = ','.join(['%1.2f' % var[i] if i < len(N) else 'nan' for i in range(k)]) + '\n'
    f.write(line)
```

## a)

```python
plt.title('means')
plt.boxplot(mean.T, tick_labels=N)
plt.ylabel(r'$\bar X_n$')
plt.xlabel(r'n')
```

As we can see here, the dispersion of $\bar X_n$ decreases with the increase of $n$. For $n = 250$ all points are localized in the small interval in contrast to $\bar X_n$ points related to $n = 8$, which are more spread.

## b)

As $Var(\bar X_n) = \frac{\sigma ^ 2}{n}$ and $\sigma = \lambda = 1$, 

$Var(\bar X_n) = \frac{1}{n}$

```python
plt.title('vars')
plt.plot(1/np.array(N), var, '.')
plt.plot([0, 1], [0, 1], '--')
plt.ylabel(r'$Var(\bar X_n)$')
plt.xlabel(r'$\frac{1}{n}$')
```
