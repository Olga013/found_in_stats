## Pen and paper ex.1

Let $P(A)$ - is a probability that at least 2 students share the same birthday.

Let $P(\overline{A})$ = 1 - $P(A)$ - is a probability that nobody shares the birthday.

$P(\overline{A}) = \frac{365 \cdot 364 \cdot 363 ... \cdot 306}{365 \cdot 365 \cdot 365 ... \cdot 365} = \frac{365 !}{305! \cdot 365^{60}}$

$P(A) = 1 - \frac{365 !}{305! \cdot 365^{60}} = 0.994123$

```python
import math
p_a = 1 - math.factorial(365)/math.factorial(305)/365**60
print('Resulting probability: %f' % (p_a))
```

## Pen and paper ex.3

$F_X(x) = \left\{
\begin{array}{l}
1 - e^{-x},\quad x \ge 0 \\
0,\quad x < 0
\end{array}
\right.
$

<br>

**a)** Find PDF $f_X(x)$ - ?

$f_X(x) = F_X^{'}(x) = \left\{
\begin{array}{l}
e^{-x},\quad x \ge 0 \\
0,\quad x < 0
\end{array}
\right.$

<br>

**b)** Show it is a valid PDF:

$1)$ $\int_{-\infty}^{+\infty} f_X(x)\,dx = 1$

$2)$ $f_X(x) \ge 0,\quad \forall x$

<br>

$1)$ $\int_{-\infty}^{+\infty} f_X(x)\,dx = \underset{\text{ = 0}}{\int_{-\infty}^{0} f_X(x)\,dx } + \int_{0}^{+\infty} f_X(x)\,dx = \int_{0}^{+\infty} e^{-x}\,dx = \left. -e ^{-x} \right |_{0}^{+\infty} = \underset{\text{ = 0}}{\left. -e ^{-x} \right |_{x \to +\infty}} + 1 = 1$

$2)$ As $e^{-x} > 0,\quad \forall x $ 

Therefore $f_X(x) \ge 0,\quad \forall x$

```python
import matplotlib.pyplot as plt
import numpy as np
n = 100
x = np.arange(-10, 10, 0.1, dtype=float)
f_x = np.exp((-1) * x)
f_x[x < 0] = 0
plt.plot(x, f_x)
plt.xlabel('x')
plt.ylabel('$f_X(x)$')
```

## Pen and paper ex.4

$f_{XY}(x,y) = \left\{
\begin{array}{l}
c,\quad 0 \le x \le y \le 1 \\
0,\quad elsewhere
\end{array}
\right.$

<br>

**a)** Value of a constant $c$ - ?

As $f_{XY}(x,y)$ is a PDF function it should satisfy the equation:

$\int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} f_{XY}(x,y)\,dx dy = 1$

$\int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} c\,dx dy = 1$

$\int_{0}^{1} \int_{0}^{1} c\,dx dy = 1$

$c \int_{0}^{1} \,dx \int_{0}^{1} \,dy = 1$

$c \cdot 1 \cdot 1 = 1$

$c = 1$

<br>

**b)** $f_X(x)$ - ? $f_Y(y)$ - ? Are $X$, $Y$ independent?

$f_{XY}(x,y) = \left\{
\begin{array}{l}
1,\quad 0 \le x \le y \le 1 \\
0,\quad elsewhere
\end{array}
\right.$

<br>

$f_X(x) = \int_{}{}f_{XY}(x,y) \, dy = \int_{-\infty}^{0} 0 \,dy + \int_{0}^{1} 1 \,dy + \int_{0}^{+\infty} 0 \,dy = 1,\quad 0 \le x \le 1$

$f_X(x) = 0,\quad elsewhere$

<br>

$f_Y(y) = \int f_{XY}(x,y) \, dx = \int_{-\infty}^{0} 0 \,dx + \int_{0}^{1} 1 \,dx + \int_{0}^{+\infty} 0 \,dx = 1,\quad 0 \le y \le 1$

$f_Y(y) = 0,\quad elsewhere$

<br>

So,

$f_{XY}(x,y) = f_X(x) * f_Y(y)$ is True,

Therefore, $X$ and $Y$ - independent.

$Z = min\{X, Y\}$

$f_Z(z)$ - ?

<br>

$F_Z(z) = P(Z \le z) = 1 - P(Z > z)$, where $z$ is parameter

$f_Z(z) = F_Z^{'}(z)$

$P(Z > z) = P(\{x, y\}: z(x, y) > z) = \int \int_{A_z} f_{XY}(x,y) = \int \int_{A_z} f_{X}(x) f_{Y}(y)$

**1)** $ 0 \le z \le 1$

$min(x, y) > z : \left\{
\begin{array}{l}
x,\quad y > x > z \\
y,\quad x \ge y > z
\end{array}
\right.
$

$P(Z > z) = \int \int_{z \in [0, 1]} f_{X}(x) f_{Y}(y) =  \int_{z}^{1} \, dx \int_{z}^{1} 1 \, dy = \int_{z}^{1} \, dx \, (1 - z) = (1 - z)^2$

**2)** $ z < 0 $

$P(Z > z) = 1$

**3)** $ z > 1 $

$P(Z > z) = 0$

<br>

As $P(Z \le z) = 1 - P(Z > z)$:


$F_Z(z) = P(Z \le z) = \left\{

0,\quad z < 0
\end{array}
\right.
$

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#a function to get an analytical solution from the ex. above
def get_analytical_solution(z):
    result = np.zeros(len(z))
    for i in range(len(z)):
        result[i] = (2 - 2 * z[i]) if (z[i] >= 0 and z[i] <= 1) else 0
    return result
```

```python
#i.)
## IMPORT DATA FROM Input.txt
inputdata = np.loadtxt('../data/Input.txt')
#print(inputdata)
inputdata = [int(item) for item in inputdata]

## SET RANDOM SEED
np.random.seed(inputdata[0])
```

```python
#Get a number of samples from an input file
n = inputdata[1]
```

```python
#Sample from the uniform distribution
x = np.random.uniform(low=0.0, high=1.0, size=n)
y = np.random.uniform(low=0.0, high=1.0, size=n)
z = np.minimum(x, y)
```

```python
#Get an analytical solution
z_a = np.arange(-0.5, 1.5, 0.01, dtype=float)
fz_a = get_analytical_solution(z_a)
```

```python
#Read a list of the values for the random variable Z
zsamples = np.genfromtxt('../data/ZSamples.txt', delimiter=',')[: -1]
```

```python
#Check if our sampled Z values are the same as Z located in the file
assert np.isclose(z, zsamples, atol=1e-02).all()
```

```python
#Plot both simulated Z and the analytical solution
plt.hist(z, density=True, label='simulation')
plt.plot(z_a, fz_a, label='analytical')
plt.legend()
plt.show()
```
